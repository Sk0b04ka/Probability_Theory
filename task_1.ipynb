## Геометрическая вероятность: «дождь и клумба»

### Условие задачи

Рассматривается квадратный двор размером $3 \times 3$.
В центре двора расположена ромбовидная клумба с углами $90^\circ$

Капля дождя падает равновероятно в любую точку двора.
Требуется найти вероятность того, что капля попадет на клумбу.

---

### Теоретический расчет

Площадь двора:
$S_{\text{двор}} = 3 \cdot 3 = 9$

Клумба представляет собой ромб с диагоналями длины $2$ и $2$.
Площадь ромба:
$S_{\text{клумба}} = \frac{d_1 d_2}{2} = \frac{2 \cdot 2}{2} = 2$

Геометрическая вероятность попадания капли на клумбу:
$P = \frac{S_{\text{клумба}}}{S_{\text{двор}}} = \frac{2}{9}$

---

### Численное моделирование

Для проверки теоретического результата используется метод Монте-Карло.
Случайные точки равномерно распределяются по всей площади двора,
и вычисляется доля точек, попавших внутрь клумбы.
При большом числе испытаний полученная вероятность
приближается к значению $\frac{2}{9}$.


```python
import numpy as np
import matplotlib.pyplot as plt

N = 20000 

x = np.random.uniform(-1.5, 1.5, N)
y = np.random.uniform(-1.5, 1.5, N)

inside_rhombus = (np.abs(x) + np.abs(y) <= 1)

estimated_probability = np.sum(inside_rhombus) / N

plt.figure(figsize=(6, 6))

plt.scatter(
    x[inside_rhombus],
    y[inside_rhombus],
    s=2,
    alpha=0.5,
    label="Попадание в клумбу"
)

plt.scatter(
    x[~inside_rhombus],
    y[~inside_rhombus],
    s=2,
    alpha=0.5,
    label="Мимо клумбы"
)

rhombus_x = [0, 1, 0, -1, 0]
rhombus_y = [1, 0, -1, 0, 1]
plt.plot(rhombus_x, rhombus_y)

yard_x = [-1.5, 1.5, 1.5, -1.5, -1.5]
yard_y = [-1.5, -1.5, 1.5, 1.5, -1.5]
plt.plot(yard_x, yard_y)

plt.gca().set_aspect("equal")
plt.title("Геометрическая вероятность: дождь и клумба")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

print("Фактически полученная вероятность:", estimated_probability)
```
