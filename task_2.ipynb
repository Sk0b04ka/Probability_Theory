## Связь классического и частотного определения вероятности
### (на примере суммы очков на двух кубиках)

### Условие задачи

Бросаются два честных игральных кубика.
Рассматривается событие $A$ — сумма выпавших очков равна $7$.

Требуется:
- найти вероятность события $A$ по классическому определению;
- смоделировать серию экспериментов;
- показать, что частотная вероятность стремится к классической
  при увеличении числа опытов.

---

### Классическое определение вероятности

При броске двух кубиков существует $36$ равновозможных исходов.

Сумма $7$ получается в следующих случаях:
$(1,6), (2,5), (3,4), (4,3), (5,2), (6,1)$

Число благоприятных исходов равно $6$.

Классическая вероятность:
$P_{\text{класс}} = \frac{6}{36} = \frac{1}{6}$

---

### Частотное определение вероятности

Пусть проведено $n$ независимых опытов.
Обозначим через $k_n$ число появлений события $A$.

Частотная вероятность определяется как:
$P_{\text{част}}(n) = \frac{k_n}{n}$

Согласно закону больших чисел, при $n \to \infty$ выполняется:
$P_{\text{част}}(n) \to P_{\text{класс}} = \frac{1}{6}$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 20000 
target_sum = 7

dice1 = np.random.randint(1, 7, N)
dice2 = np.random.randint(1, 7, N)

sums = dice1 + dice2
success = (sums == target_sum)


cumulative_successes = np.cumsum(success)
trials = np.arange(1, N + 1)

frequency_probability = cumulative_successes / trials

classical_probability = 1 / 6

plt.figure(figsize=(8, 5))

plt.plot(trials, frequency_probability, label="Частотная вероятность")
plt.axhline(
    y=classical_probability,
    linestyle="--",
    label="Классическая вероятность (1/6)"
)

plt.xlabel("Число опытов")
plt.ylabel("Вероятность")
plt.title("Сходимость частотной вероятности к классической")
plt.legend()
plt.grid(True)

plt.show()

print("Частотная вероятность при N =", N, ":", frequency_probability[-1])
print("Классическая вероятность:", classical_probability)
```
