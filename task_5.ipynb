## 1. Биномиальное распределение

**Условие задачи:**  
На фабрике производятся партии микрочипов. Вероятность того, что случайно выбранный микрочип окажется бракованным, равна $p = 0.2$.  
Выборка состоит из $n = 12$ независимых микрочипов.  
Пусть дискретная случайная величина $X$ — число бракованных микрочипов в партии из 12 штук.

- составить закон распределения ;  
- вычислить математическое ожидание , дисперсию, среднее квадратическое отклонение и моду распределения;  
- построить многоугольник распределения.




```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

def mode(xs, ps):
    return xs[np.argmax(ps)]

def draw(xs, ps, title):
    plt.figure(figsize=(6,4))
    plt.plot(xs, ps, marker="o")
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("P(x)")
    plt.grid(True)
    plt.show()

# параметры
n = 12
p = 0.20

xs = np.arange(0, n+1)
ps = [math.comb(n, k) * p**k * (1-p)**(n-k) for k in xs]

df_bin = pd.DataFrame(data=ps, index=xs, columns=["P(X)"])
display(df_bin)

mean = n*p
var = n*p*(1-p)
std = np.sqrt(var)
moda = math.floor((n+1)*p)

print("Математическое ожидание =", round(mean, 4))
print("Дисперсия =", round(var, 4))
print("Среднеквадратичное отклонение =", round(std, 4))
print("Мода =", round(moda, 4))

draw(xs, ps, "Биномиальное распределение")
```



## 2. Распределение Пуассона

**Условие задачи:**  
На перекрестке за 10 минут в среднем происходит $\lambda = 3.75$ случая проезда машины на запрещающий сигнал.  
Пусть дискретная случайная величина $X$ — число таких нарушений за очередные 10 минут наблюдений.

- составить закон распределения ;  
- вычислить математическое ожидание , дисперсию, среднее квадратическое отклонение и моду распределения;  
- построить многоугольник распределения.


```python
from scipy.stats import poisson

lmb = 3.75
xs = np.arange(0, 11)

ps = poisson.pmf(xs, lmb)

df = pd.DataFrame(data=ps.reshape(1, -1), columns=xs, index=["P(X)"])
display(df)

print("Мат. ожидание =", poisson.mean(lmb))
print("Дисперсия =", poisson.var(lmb))
print("Среднеквадратичное отклонение =", round(poisson.std(lmb), 4))
print("Мода =", int(lmb))

draw(xs, ps, "Распределение Пуассона")
```

## 3. Геометрическое распределение

**Условие задачи:**  
Вероятность того, что покупатель на сайте оформит заказ, равна $p = 0.12$.  
Пусть дискретная случайная величина $X$ — номер по счету попытки (визита), на которой впервые будет сделан заказ.

- составить закон распределения ;  
- вычислить математическое ожидание , дисперсию, среднее квадратическое отклонение и моду распределения;  
- построить многоугольник распределения.


```python
p = 0.12
xs = np.arange(1, 11)
ps = [(1-p)**(k-1) * p for k in xs]

df_geo = pd.DataFrame(ps, index=xs, columns=["P(X)"])
display(df_geo)

mean = 1/p
var = (1-p)/p**2
std = np.sqrt(var)
moda = 1

print("Мат. ожидание =", round(mean, 4))
print("Дисперсия =", round(var, 4))
print("Среднеквадратичное отклонение =", round(std, 4))
print("Мода =", round(moda, 4))

draw(xs, ps, "Геометрическое распределение")
```


## 4. Гипергеометрическое распределение

**Условие задачи:**  
В коробке лежат $N = 100$ пакетов молока, из которых 9 — с истёкшим сроком годности.  
Случайным образом выбирают 5 пакетов.  
Пусть дискретная случайная величина $X$ — число просроченных пакетов молока среди выбранных.

- составить закон распределения ;  
- вычислить математическое ожидание , дисперсию, среднее квадратическое отклонение и моду распределения;  
- построить многоугольник распределения.


```python
from math import comb

N = 100
K = 9
n = 5
xs = np.arange(0, n+1)
ps = [comb(K, k) * comb(N-K, n-k) / comb(N, n) for k in xs]

df_hyper = pd.DataFrame(ps, index=xs, columns=["P(X)"])
display(df_hyper)

mean = n * K / N
var = n * K * (N-K) * (N-n) / (N**2 * (N-1))
std = np.sqrt(var)

moda = math.floor((n+1) * (K+1) / (N+2))

print("Мат. ожидание =", round(mean, 4))
print("Дисперсия =", round(var, 4))
print("Ст. отклонение =", round(std, 4))
print("Мода =", round(moda, 4))

draw(xs, ps, "Гипергеометрическое распределение")
```
