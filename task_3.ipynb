```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# ---------------------- Описание датасета ----------------------

display(Markdown(
    "### Датасет содержит химические показатели вина и целевой признак — качество вина по шкале 0–10."
))
# ---------------------- Загрузка данных ----------------------

df = pd.read_csv("WineQT.csv")

print("Размер датасета:", df.shape)
display(df.head())
display(df.info())

# Полное переименование всех признаков
df = df.rename(columns={
    "fixed acidity": "Фиксированная кислотность",
    "volatile acidity": "Летучая кислотность",
    "citric acid": "Лимонная кислота",
    "residual sugar": "Остаточный сахар",
    "chlorides": "Хлориды",
    "free sulfur dioxide": "Свободный SO2",
    "total sulfur dioxide": "Общий SO2",
    "density": "Плотность",
    "pH": "pH",
    "sulphates": "Сульфаты",
    "alcohol": "Алкоголь",
    "quality": "Качество"
})

# Удаляем столбец с идентификаторами
if "Id" in df.columns:
    df = df.drop(columns=["Id"])

display(df.head())
display(df.describe())

# 1 = хорошее вино, 0 = плохое
df["Хорошее качество"] = (df["Качество"] >= 6).astype(int)

def conditional_probs(data: pd.DataFrame, col):
    return (
        data.groupby(col, observed=True)["Хорошее качество"]
        .mean()
        .reset_index(name=f"P(Хорошее качество|{col})")
        .style.hide(axis="index")
    )


def bayes_conditional(data: pd.DataFrame, col):
    p_good = (data["Хорошее качество"] == 1).mean()
    p_bad = 1 - p_good

    rows = []
    for val in data[col].dropna().unique():
        p_val = (data[col] == val).mean()

        p_good_given_val = data.loc[data[col] == val, "Хорошее качество"].mean()
        p_bad_given_val = 1 - p_good_given_val

        p_val_given_good = (p_good_given_val * p_val) / p_good
        p_val_given_bad = (p_bad_given_val * p_val) / p_bad

        rows.append([val, p_val_given_good, p_val_given_bad])

    return pd.DataFrame(
        rows,
        columns=[col, f"P({col}|Хорошее)", f"P({col}|Плохое)"],
    ).style.hide(axis="index")

df_grouped = pd.DataFrame(df)

df_grouped["Группа алкоголя"] = pd.cut(
    df["Алкоголь"],
    bins=[5, 9, 11, 13, 16],
    labels=["5-9", "9-11", "11-13", "13+"],
)

df_grouped["Группа сахара"] = pd.cut(
    df["Остаточный сахар"],
    bins=[0, 2, 5, 10, 20],
    labels=["0-2", "2-5", "5-10", "10+"],
)

df_grouped["Группа летучей кислотности"] = pd.cut(
    df["Летучая кислотность"],
    bins=[0.1, 0.3, 0.5, 1.0, 2.0],
    labels=["0.1-0.3", "0.3-0.5", "0.5-1.0", "1.0+"],
)

df_grouped["pH-группа"] = pd.cut(
    df["pH"],
    bins=[2.5, 3.0, 3.3, 3.6, 4.0],
    labels=["2.5-3.0", "3.0-3.3", "3.3-3.6", "3.6+"],
)

display(df_grouped.head())

display(Markdown("### Вероятность хорошего качества по содержанию алкоголя"))
display(conditional_probs(df_grouped, "Группа алкоголя"))

display(Markdown("### Вероятность хорошего качества по остаточному сахару"))
display(conditional_probs(df_grouped, "Группа сахара"))

display(Markdown("### Вероятность хорошего качества по pH"))
display(conditional_probs(df_grouped, "pH-группа"))

display(Markdown("### Вероятность хорошего качества по летучей кислотности"))
display(conditional_probs(df_grouped, "Группа летучей кислотности"))

display(Markdown("### Байес: распределение алкоголя при хорошем/плохом качестве"))
display(bayes_conditional(df_grouped, "Группа алкоголя"))

display(Markdown("### Байес: распределение остаточного сахара"))
display(bayes_conditional(df_grouped, "Группа сахара"))

display(Markdown("### Байес: распределение летучей кислотности"))
display(bayes_conditional(df_grouped, "Группа летучей кислотности"))

for col in ["Группа алкоголя", "Группа сахара", "pH-группа", "Группа летучей кислотности"]:
    plt.figure(figsize=(6, 4))
    sns.barplot(
        x=col, y="Хорошее качество", data=df_grouped, estimator=np.mean, errorbar=None
    )
    plt.title(f"Вероятность хорошего качества в зависимости от {col}")
    plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Корреляции химических признаков вина")
plt.show()

display(
    Markdown(
        """
### Итоговые наблюдения:

- Чем больше **алкоголя**, тем выше вероятность хорошего качества вина.
- Повышенная **летучая кислотность** ухудшает качество.
- Очень низкий или очень высокий уровень **остаточного сахара** чаще связан с более низким качеством.
- **pH** в диапазоне 3.3–3.6 чаще встречается у лучших образцов.
"""
    )
)
```
